```python
import pandas as pd

# Loading dataset
df_raw = pd.read_excel('Dataset_for_Data_Analytics.xlsx')
df = df_raw.copy()
print(df.shape)
df.head()
```

```python
# Exploring dataset structure
df.info()
```

```python
# Statistical summary
df.describe()
```

```python
# Checking missing values
df.isnull().sum()
```

```python
# Handling missing values
df['CouponCode'] = df['CouponCode'].fillna('No Coupon')

# Re-checking missing values
df.isnull().sum()
```

```python
# Checking duplicates
print("Duplicate OrderIDs:", df['OrderID'].duplicated().sum())
print("Duplicate full rows:", df.duplicated().sum())

# Handling duplicates
df = df.drop_duplicates(subset='OrderID')
df = df.drop_duplicates()

# Re-checking duplicates
print("\nDuplicate OrderIDs after cleaning:", df['OrderID'].duplicated().sum())
print("Duplicate full rows after cleaning:", df.duplicated().sum())
```

```python
# Checking date and number formatting
print(df['Date'].dtype)
print((df['UnitPrice'].round(2) != df['UnitPrice']).sum())
print((df['TotalPrice'].round(2) != df['TotalPrice']).sum())
```

```python
# Handling number formatting
df['TotalPrice'] = df['TotalPrice'].round(2)

# Handling date format
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')

# Re-checking date and number formatting
print(df['Date'].head())
print("\nUnitPrice mismatches:", (df['UnitPrice'].round(2) != df['UnitPrice']).sum())
print("TotalPrice mismatches:", (df['TotalPrice'].round(2) != df['TotalPrice']).sum())
```

```python
# Checking text formatting
text_cols = ['Product', 'ShippingAddress', 'PaymentMethod', 'OrderStatus', 'ReferralSource', 'CouponCode']
for col in text_cols:
    print(col, ":", df[col].unique())
    print()
```

```python
# Exporting cleaned dataset
df.to_excel('cleaned_dataset.xlsx', index=False)
```
